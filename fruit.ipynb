```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Class"]=le.fit_transform(df["Class"])
df
```

```python
y=df['Class']
y
```

```python
x=df.drop(['Class'],axis=1)
x
```

```python
from sklearn.preprocessing import StandardScaler
SD=StandardScaler()
x=SD.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
LG = LogisticRegression()
LG.fit(x_train, y_train)
```

```python
y_pred1 = LG.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred1, labels=[1,0]))
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, y_pred1)
print(accuracy)
print(metrics.classification_report(y_test, y_pred1, labels=[1, 0]))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN= KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
KNN.fit(x_train, y_train)
```

```python
y_pred2 = KNN.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred2, labels=[1,0])
ac = accuracy_score(y_test, y_pred2)
print(cm)
print(ac)
```

```python
print(metrics.classification_report(y_test, y_pred2, labels=[1, 0]))
```
